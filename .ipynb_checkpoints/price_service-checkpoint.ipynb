```python
QUANTILE_ARM = {0: '5%', 1: '50%', 2: '95%'}

trades = (pd.read_csv(DIR + TRADES)
    .assign(year_week = lambda x: x['year'].astype(str) + '-' + x['week_num'].astype(str))
    .assign(year_week = lambda x: pd.to_datetime(x['year_week'] + '-1', format="%Y-%W-%w"))
)

def display_result(result):
  global trades
  global QUANTILE_ARM

  result_plot = (result.assign(year_week = lambda x: pd.to_datetime(x['dates']).dt.strftime("%Y-%W"))
        .assign(year_week = lambda x: pd.to_datetime(x['year_week'] + '-1', format="%Y-%W-%w"))
        # .merge(trades, on = ['SKU', 'year_week'], how = 'left')
        .assign(revenue = lambda x: (x['price'] - x['cost_price']) * x['bought'])
        .assign(revenue = lambda x: np.where(x['revenue'] == 0.0, np.nan, x['revenue']))
  )

  fig, axes = plt.subplots(3, 2, figsize=(10, 6))
  axes = iter(axes.flatten())
  ax = next(axes)
  sns.histplot(data = result_plot.query("~revenue.isna()"), x = 'revenue',  ax = ax), ax.set_title('revenue per SKU distribution')

  ax = next(axes)
  sns.histplot(data = result_plot, x = 'price', hue = 'bought', ax = ax), ax.title.set_text('price distribution')

  ax = next(axes)
  df_plot = result_plot.groupby(['SKU', 'bought']).agg({'user_id':'count'}).reset_index()
  sns.histplot(data = df_plot, x = 'user_id', hue = 'bought', ax = ax), ax.set_title('SKU num purchase distribution')

  ax = next(axes)
  df_plot = result_plot.groupby(['user_id', 'bought']).agg({'SKU':'count'}).reset_index()
  sns.histplot(data = df_plot, x = 'SKU', hue = 'bought', ax = ax), ax.set_title('different SKU by each user distribution')

  ax = next(axes)
  boxplot = (result.query("revenue != 0")
                   .assign(markup_perc = lambda x: x['revenue'] / x['cost_price'] * 100)
                   .assign(arm = lambda x: x['arm'].map(QUANTILE_ARM)))
  sns.boxplot(data = boxplot, x = 'markup_perc', hue = 'arm', ax = ax, flierprops=dict(marker='o', markersize=1))
  ax.set_title('markup by quantile distribution'), ax.legend_.set_title("Quantile")

  ax = next(axes)
  barplot = boxplot.groupby('arm').agg({'bought':'sum'}).reset_index()
  sns.barplot(data =  barplot, x = 'bought', hue = 'arm', ax = ax), ax.set_title('Purchase amount by price per SKU quantile'), ax.legend_.set_title("Quantile")

  plt.suptitle(result_plot['dates'].max(), fontsize=16)
  plt.tight_layout()
  plt.show()

  df_plot =(result_plot.query("revenue < 0")
                      .groupby('SKU')
                      .agg(loss_price = ('revenue', 'sum'), num_pruchase = ('user_id', 'count'))
                      .assign(loss = lambda x: x['loss_price'] * x['num_pruchase'])
                      .sort_values('loss', ascending = True))
  display(df_plot)
  return fig, axes
```

```python
# save data
class recorder():
  def __init__(self, path):
    self.path = path

  def record(self, df):
    if not os.path.isfile(self.path):
      write(self.path, df)
    else:
      write(self.path, df, append=True)
    pass
```

```python
# Server session class to POST and GET data from market server
class ServerSession():
  def __init__(self,
               UUID = 'e1dc7312137340baa428c78f282d00dc',
               path = None,
               ):
    self.UUID = UUID
    self.recorder = recorder(path)
    # endpoints
    self.URL_TASK_DATA_GET = 'https://lab.karpov.courses/hardml-api/project-1/task/{uuid}/data'
    self.URL_BEGIN_DATA = 'https://lab.karpov.courses/hardml-api/project-1/task/{uuid}/begin'
    self.URL_TASK_RESULT_POST = 'https://lab.karpov.courses/hardml-api/project-1/task/{uuid}/result/'
    self.URL_TASK_RESULT_GET = 'https://lab.karpov.courses/hardml-api/project-1/task/{uuid}/result'

  def start(self):
    req = requests.post(self.URL_BEGIN_DATA.format(uuid=self.UUID))
    if req.json()['status'] == 'ok':
      print(req)
      pass
    else:
      raise Exception('status not ok')

  def get_purchase_request(self):
    response = requests.get(self.URL_TASK_DATA_GET.format(uuid=UUID))
    try:
      df = pd.read_json(StringIO(response.json()))
      return df
    except Exception as e:
      self.end_session(response)
      return None

  def post_offer(self, df):
    # обратите внимание, что батч данных вам надо подготовить к загрузке определенным образом:
    # df.to_json(orient='records')
    # Обратите внимание, что выходной файл должен выглядеть следующий образом:
    # [{"dates":"2019-12-01","SKU":23198,"user_id":16625,"price":1000.0}]
    req = requests.post(self.URL_TASK_RESULT_POST.format(uuid=self.UUID),
                            data=df.to_json(orient='records'))
    result = self.offer_result()
    self.recorder.record(result)
    return result

  def offer_result(self):
    response = requests.get(self.URL_TASK_RESULT_GET.format(uuid=self.UUID))
    # Чтобы перевести полученный ответ в pd.DataFrame
    df = pd.read_json(StringIO(response.json()))
    return df

  def end_session(self,response):
    # 'batch processing finished ' - окончание работы
    print(response.json()['status'])
    return None
```

```python
# define main MultipleArmBandit class
class Strategy:
    def __init__(self, n_arms: int):
        self.n_arms = n_arms
        self.n_iters = 0
        self.arms_states = np.zeros(n_arms)
        self.arms_actions = np.zeros(n_arms)

    def flush(self):
        self.n_iters = 0
        self.arms_states = np.zeros(self.n_arms)
        self.arms_actions = np.zeros(self.n_arms)

    def update_reward(self, arm: int, reward: int):
        self.n_iters += 1
        self.arms_states[int(arm)] += reward
        self.arms_actions[int(arm)] += 1

    def choose_arm(self):
        raise NotImplementedError
# define Thompson strategy algorytm
class Thompson(Strategy):

    def __init__(self, n_arms: int):
      # your code here
      super().__init__(n_arms)
      self.alfas = np.ones(n_arms)
      self.betas = np.ones(n_arms)

    def update_reward(self, arm: int, reward: int):
      # your code here
      super().update_reward(arm, reward)
      if reward > 0:
        self.alfas[int(arm)] += 1
      else:
        self.betas[int(arm)] += 1

    def choose_arm(self):
    # your code here
        betas = [np.random.beta(self.alfas[arm],self.betas[arm]) for arm in range(self.n_arms)]
        arm = np.argmax(betas)
        return arm
# define Epsilon greedy trategy algorytm
class EpsGreedy(Strategy):

    def __init__(self, n_arms: int, eps: float = 0.1):
        super().__init__(n_arms)
        self.eps = eps

    def choose_arm(self):
        if random.random() < self.eps:
            return random.randint(0, self.n_arms - 1) # choose random arm
        else:
            return np.argmax(self.arms_states / self.arms_actions) # choose the best arm
# define class to contain bandits for each user_id
class MultiArmBandit():
  def __init__(self,
               N_arm,
               ):
    self.N = N_arm
    self.users = {}

  def add_user(self, user_id):
    self.users[user_id] = EpsGreedy(self.N)

  def get_arm(self, user_id):
    if user_id not in self.users:
      self.add_user(user_id)
    return self.users[user_id].choose_arm()

  def update_reward(self, user_id, arm, reward):
    self.users[user_id].update_reward(arm, reward)

  def flush(self):
    for user in self.users:
      self.users[user].flush()
```

```python
PATH =  DIR + f'session_purchases'
PREDICTIONS = DIR + 'predictions_margine_lstm_quantille.parquet'
SELECT = ['dates', 'SKU', 'user_id', 'price']
price_predictions = pd.read_parquet(PREDICTIONS)
COLUMNS = ['price_pred_q5', 'price_pred_q50', 'price_pred_q95']
SELECT_OFFER = ['dates', 'SKU', 'user_id', 'arm', 'cost_price']


if os.path.exists(PATH):
    os.remove(PATH)
    print(f"File {PATH} has been removed successfully.")
else:
    print(f"File {PATH} does not exist.")

# start session with server
session = ServerSession(path = PATH)
session.start()

# get price predictions quantiles
price_predictions = pd.read_parquet(PREDICTIONS)
# create bandit for each user_id
bandit = MultiArmBandit(len(COLUMNS))

history = pd.DataFrame()

while True:
  df_current = session.get_purchase_request()
  if df_current is None:
    break
  else:
    print(df_current['dates'].drop_duplicates().to_list())
    # calculate offer
    offer = (df_current.assign(year_week = lambda x: pd.to_datetime(x['dates']).dt.strftime("%Y-%W"))
                       .assign(year_week = lambda x: pd.to_datetime(x['year_week'] + '-1', format="%Y-%W-%w"))
                       .merge(price_predictions, on = ['year_week', 'SKU'], how = 'left')
                       .assign(arm = lambda x: x['user_id'].apply(lambda user: bandit.get_arm(user)))
                       .assign(price = lambda x: x.apply(lambda y: y[COLUMNS[y['arm']]], axis = 1))
                       )
    # post offer get bought result
    result = session.post_offer(offer[SELECT])
    # update_reward for all bandits

    result = (result.merge(offer[SELECT_OFFER], on = ['dates', 'SKU', 'user_id'], how = 'left')
                    .assign(revenue = lambda x: (x['price'] - x['cost_price']) * x['bought'])
                    .query("~revenue.isna()"))
    history = pd.concat([history, result])

    # update_reward
    (result.apply(lambda x: bandit.update_reward(user_id = x['user_id'],
                                                 arm = x['arm'],
                                                 reward = x['revenue']), axis = 1))

    display_result(history)
    plt.show()
```

```python
from matplotlib import ticker
df_plot = (history.groupby(['dates', 'arm'])
       .agg({'bought':'sum'})
       .assign(all_purchases = lambda x: x.groupby(['dates']).transform('sum')).reset_index()
       .assign(frequency = lambda x: x['bought'] / x['all_purchases'])
       .assign(arm = lambda x: x['arm'].map(QUANTILE_ARM))
)
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(data = df_plot, x = 'dates', y = 'frequency',hue = 'arm')
ax.xaxis.set_major_locator(ticker.MaxNLocator(7)), ax.set_title('Multi-arm bandit training process'), ax.legend_.set_title("Quantile of markup for price per SKU")
ax.set_ylabel('frequency of purchase')
plt.tight_layout()
plt.show()
```

```python
# get arms state from each user
objects = np.array(tuple(bandit.users.items()))[:,1]

def get_param(obj):
    return obj.arms_states

vectorized_get_param = np.vectorize(get_param)
vectorized_get_arms_states = np.frompyfunc(vectorized_get_param, 1, 1)
arms_states = np.stack(vectorized_get_arms_states(objects))
arms_states = pd.DataFrame(arms_states, columns = COLUMNS)
```

```python
# plot arms state distribution
fig, ax = plt.subplots(figsize=(8, 4))

sns.boxplot(pd.melt(arms_states), x = 'value', ax = ax, hue = 'variable', flierprops=dict(marker='o', markersize=3))
ax.set_title('Arms state distribution')
# ax.set_xscale('log')
plt.show()
```
